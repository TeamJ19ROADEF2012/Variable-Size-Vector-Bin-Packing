```python
from itertools import product

from pandas import DataFrame

from vsvbp.generator import generator, correlated_capacities, unif_bin
from vsvbp.heuristics import bfd_item_centric, bfd_bin_centric, bin_balancing
from vsvbp.measures import *

min_fill = .9
rem_cons = .8
correlated_items = False
dev = 0.2
rate = 0.25
seed = 0
num_bins = 10
num_resources = 2
num_instances = 100

sbin_balancing = lambda *args: bin_balancing(*args, single=True)

heuristics = [
    ( 'nothing', bfd_item_centric, do_nothing, do_nothing),
    ( 'shuff1', bfd_item_centric, shuffleItemsOnce, shuffleBinsOnce),
    ( '1/C', bfd_item_centric, staticItemsOneOverC, staticBinsOneOverC),
    ( '1/R', bfd_item_centric, staticItemsOneOverR, staticBinsOneOverR),
    ( 'R/C', bfd_item_centric, staticItemsROverC, staticBinsROverC),
    ( 'ic_shuff', bfd_item_centric, shuffleItems, shuffleBins),
    ( 'ic_dyn_1/C', bfd_item_centric, dynamicItemsOneOverC, dynamicBinsOneOverC),
    ( 'ic_dyn_1/R', bfd_item_centric, dynamicItemsOneOverR, dynamicBinsOneOverR),
    ( 'ic_dyn_R/C', bfd_item_centric, dynamicItemsROverC, dynamicBinsROverC),
    ( 'bc_shuff', bfd_bin_centric, shuffleItems, shuffleBins),
    ( 'bc_dyn_1/C', bfd_bin_centric, dynamicItemsOneOverC, dynamicBinsOneOverC),
    ( 'bc_dyn_1/R', bfd_bin_centric, dynamicItemsOneOverR, dynamicBinsOneOverR),
    ( 'bc_dyn_R/C', bfd_bin_centric, dynamicItemsROverC, dynamicBinsROverC),
    ( 'bb_nothing', bin_balancing, do_nothing, do_nothing),
    ( 'bb_shuff1', bin_balancing, shuffleItemsOnce, shuffleBinsOnce),
    ( 'bb_shuff', bin_balancing, shuffleItems, shuffleBins),
    ( 'bb_st_1/C', bin_balancing, staticItemsOneOverC, staticBinsOneOverC),
    ( 'bb_dyn_1/C', bin_balancing, dynamicItemsOneOverC, dynamicBinsOneOverC),
    ( 'bb_st_1/R', bin_balancing, staticItemsOneOverR, staticBinsOneOverR),
    ( 'bb_dyn_1/R', bin_balancing, dynamicItemsOneOverR, dynamicBinsOneOverR),
    ( 'bb_st_R/C', bin_balancing, staticItemsROverC, staticBinsROverC),
    ( 'bb_dyn_R/C', bin_balancing, dynamicItemsROverC, dynamicBinsROverC),
    ( 'sbb_nothing', sbin_balancing, do_nothing, do_nothing),
    ( 'sbb_shuff1', sbin_balancing, shuffleItemsOnce, shuffleBinsOnce),
    ( 'sbb_shuff', sbin_balancing, shuffleItems, shuffleBins),
    ( 'sbb_st_1/C', sbin_balancing, staticItemsOneOverC, staticBinsOneOverC),
    ( 'sbb_dyn_1/C', sbin_balancing, dynamicItemsOneOverC, dynamicBinsOneOverC),
    ( 'sbb_st_1/R', sbin_balancing, staticItemsOneOverR, staticBinsOneOverR),
    ( 'sbb_dyn_1/R', sbin_balancing, dynamicItemsOneOverR, dynamicBinsOneOverR),
    ( 'sbb_st_R/C', sbin_balancing, staticItemsROverC, staticBinsROverC),
    ( 'sbb_dyn_R/C', sbin_balancing, dynamicItemsROverC, dynamicBinsROverC),  
]

problems = [generator(num_bins, num_resources, min_fill, unif_bin, seed+i, rem_cons=rem_cons,
                      proc_rate=rate)
             for i in range(num_instances)]


results = []
for name, packer, item_measure, bin_measure in heuristics:
    result = {'num_bins': num_bins, 'num_resources': num_resources, 'heuristic' : name, 'sp_packed' : 0.0, 'n_success' : 0, 
              'total' : 0}
    results.append(result)
    for problem in problems:
        problem.empty()
        ret = packer(problem.items[:], problem.bins[:], item_measure, bin_measure)
        result['total'] += 1
        result['sp_packed'] += float(len(problem.items)-len(ret))/len(problem.items)
        if not ret:
            result['n_success'] += 1

results_df = DataFrame(results)
results_df.describe()
```

```python
from matplotlib import pyplot as plt, rcParams

#rcParams['savefig.dpi'] = 108

fig = plt.figure()
ax = fig.add_subplot(111)

results_df['pivot_idx'] = results_df['num_bins'].apply(str) + '_' + results_df['num_resources'].apply(str)

results_df.pivot('pivot_idx', 'heuristic', 'n_success').plot(ax=ax, kind='bar')

ax.set_title('Total Successes')
ax.set_xlabel('#bins_#resources')
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=0)
plt.show()
```
